```python
#Splitting data in Target and feature column
X  = data.drop(columns = "HeartDisease" , axis = 1)
Y = data["HeartDisease"]
```

```python
X
```

```python
Y
Y.dropna()
```

```python
# 1 --> Diseased
# 0 --> Healthy

#Here data is imbalaned
from imblearn.over_sampling import SMOTE hub

# Create a SMOTE object
smote = SMOTE(random_state=0)

# Resample the dataset with SMOTE
X_resampled, y_resampled = smote.fit_resample(X, Y)
```

```python
X_resampled
```

```python
y_resampled
```

```python
y_resampled.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , Y_train , Y_test = train_test_split(X_resampled ,y_resampled,test_size = 0.3 , stratify= y_resampled , random_state=3)
```

```python
from sklearn.linear_model imp
ort LogisticRegression
from sklearn.metrics import  accuracy_score

model = LogisticRegression()

model.fit(X_train,Y_train)
```

```python
X_test_prediction = model.predict(X_test)

test_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
test_accuracy
```

```python
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
training_accuracy
```

```python
#Checking train and test accuracy they should be almost same if
#if accuracy on training is larger and test is low this mean model is overfitted ie overlearn training data
#
```

```python
y_pred1 = model.predict(X_test)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate precision, recall, and F1-score
precision = precision_score(Y_test, y_pred1)
recall = recall_score(Y_test, y_pred1)
f1 = f1_score(Y_test, y_pred1)
```

```python
print("Pre " ,precision)
print("re" ,recall)
print("f" ,f1)
```

```python
Age = data["AgeCategory"]

# Create a density plot (Kernel Density Estimate) to visualize the distribution
plt.figure(figsize=(15, 10))
sns.histplot(Age, kde=True, color='skyblue')

# Set labels and title
plt.xlabel('Ages')
plt.ylabel('Density')
plt.title('Distribution of Ages')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.histplot(Age, kde=True, color='skyblue')

# Calculate mean, median, and mode
mean_age = np.mean(Age)
median_age = np.median(Age)
mode_age = float(Age.mode())  # Mode may return multiple values, so convert to float for a single value

# Display mean, median, and mode on the plot
plt.axvline(mean_age, color='red', linestyle='dashed', label=f'Mean: {mean_age:.2f}')
plt.axvline(median_age, color='green', linestyle='dashed', label=f'Median: {median_age}')
plt.axvline(mode_age, color='purple', linestyle='dashed', label=f'Mode: {mode_age}')

# Set labels and title
plt.xlabel('Ages')
plt.ylabel('Density')
plt.title('Distribution of Ages')
plt.legend()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=Age, color='skyblue')

# Set labels and title
plt.xlabel('Ages')
plt.title('Boxplot of Age Distribution')

# Show the plot
plt.show()
```
